## 1. Clean your Datasets

For the first part of your project, you will apply the same data cleaning steps to your "data/control.txt" and "data/treatment.txt" files. By the end of this step, you should have two lists of floats that match the numbers you have in each respective file.

You will implement your `clean_data` function in your `cleaner.py` module and import the module in the code-block below.

```python
# TODO: import all the packages that you will need for this notebook here
import statistics as stats
import matplotlib.pyplot as plt 
import cleaner
```

```python
# TODO: read your text files into your code using file I/O

control_file = open('data/control.txt')
treatment_file = open('data/treatment.txt')

control_data = control_file.readlines()
treatment_data = treatment_file.readlines()

# TODO: clean both list objects that you created above using the 'clean_data' function

clean_control_file = cleaner.clean_data(control_data)
clean_treatment_file = cleaner.clean_data(treatment_data)

print('Clean Control Data: ', clean_control_file)
print(f'\nClean Treatment Data: ', clean_treatment_file)
```

## 2. Explore your Datasets

Now that you have your list objects created and cleaned, let's explore the descriptive statistics of these two files and make subsequent observations on our data.

```python
# TODO: Calculate the standard deviation of both the control and treatment group

control_SD = stats.stdev(clean_control_file)
treatment_SD = stats.stdev(clean_treatment_file)

# TODO: print your findings

print(f'Control Group Standard Deviation: ', f'{control_SD:.2f}')
print(f'Treatment Group Standard Deviation', f'{treatment_SD:.2f}')
```

```python
# TODO: Calculate the mean of both the control and treatment group
control_mean = stats.mean(clean_control_file)
treatment_mean = stats.mean(clean_treatment_file)

# TODO: print your findings
print(f'Control Mean: ', f'{control_mean:.2f}')
print(f'Treatment Mean', f'{treatment_mean:.2f}')
```

```python
# TODO: Are there any outliers in your control group? (use 2 * stdev to find these outliers)
outliers_control = []
for outlier in clean_control_file:
    if outlier > control_mean + 2* control_SD or outlier < control_mean - 2* control_SD: #We're defining outliers to be anything that's more than 2 standard deviations above or below the mean. 
        outliers_control.append(outlier)
outliers_control_sorted = sorted(outliers_control)
print(f'\nSorted Outliers Control Data: ', outliers_control_sorted)
print(f'\nOutliers: ', outliers_control)
```

```python
# TODO: Are there any outliers in your treatment group?
outliers_treatment = []
for outlier in clean_treatment_file:
    if outlier > treatment_mean + 2* treatment_SD or outlier < treatment_mean - 2* treatment_SD: 
        outliers_treatment.append(outlier)
treatment_sorted_data = sorted(outliers_treatment)
print(f'\nOutliers Treatment Data: ', outliers_treatment)
print(f'\nSorted Outliers Treatment Data: ', treatment_sorted_data) 

```

## 3. Create Data Visualizations

For this final step, generate data visualizations to view the dispersion of both of your groups.

Make sure that the two visualizations you create are on the same figure. 

```python
# # TODO: create a data visualization to view differences in dispersion and mean

fig,ax = plt.subplots(1,2,facecolor='wheat', sharex=True,figsize=(11,3))
ax[0].set_title("Control Group")
ax[0].set_facecolor('lavenderblush')
ax[0].set_xlabel("Viewing_Time_Minutes"), ax[1].set_xlabel("Viewing_Time_Minutes")
ax[0].hist(clean_control_file, histtype='barstacked', linewidth=2, edgecolor='paleturquoise', color = 'lightseagreen')
ax[1].set_facecolor('mistyrose')
ax[1].set_title("Treatment Group")
ax[1].hist(clean_treatment_file, histtype= 'barstacked', color='cornflowerblue', edgecolor='darkslateblue', linewidth=2)

plt.show()
```
