```python
import pandas as pd 
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

```

```python
# TODO: implement your code here...


control_file_dataframe = pd.read_csv('data/control.txt')
treatment_file_dataframe = pd.read_csv('data/treatment.txt')
Control_summary_stats = control_file_dataframe['minutes'].describe()


print(f'Control Data Descriptive Stats: ', f"\n\tMean: ", f"{Control_summary_stats['mean']:.2f}", f'\n\tSample Standard Deviation: ' , f"{Control_summary_stats['std']:.2f}")
control_data_range = Control_summary_stats['max'] - Control_summary_stats['min']
print(f"\nControl Group Data Range: ", control_data_range)
Treatment_summary_stats = treatment_file_dataframe['minutes'].describe()
print(f'\nTreatment Data Descriptive Stats: ', f'\n\tMean: ', f"{Treatment_summary_stats['mean']:.2f}", f'\n\tSample Standard Deviation: ', f"{Treatment_summary_stats['std']:.2f}")
treatment_data_range = Treatment_summary_stats['max'] - Treatment_summary_stats['min']
print(f'\nTreatment Group Data Range: ', f'{treatment_data_range:.2f}')
```

```python

fig, ax = plt.subplots(1, 2, figsize=(13,3))
legend_data = ['Control Group', 'Treatment Group']
sns.swarmplot(data=treatment_file_dataframe, x='minutes',palette='hls', hue = 10, legend= False, ax = ax[0]).set_title('Treatment Group Data')
sns.swarmplot(data = control_file_dataframe, x = 'minutes', palette='bright', hue = 10, legend=False, ax = ax[1]).set_title('Control Group Data')
sns.set_style('darkgrid')

plt.show()
```

```python
t_test_result = stats.ttest_ind(control_file_dataframe, treatment_file_dataframe)
print(f't-test Result: ', t_test_result)

print(f'\nT-Test Statistic: ',  t_test_result.statistic)
print(f'\nT-Test P-Value: ', t_test_result.pvalue)
print(f'\nT-test Confidence Interval: ', t_test_result.confidence_interval())

```
